```python
import numpy as np;
print("1,Create a numpy array of size 2*2");
array_2_by_2=np.array([[1,2],[3,4]]);
print("2*2 Array:");
print(array_2_by_2);
print("Shape:",(array_2_by_2).shape)
print()
```

```python

random_3_by_4=np.random.rand(3,4);
print(random_3_by_4);
```

```python
ones_array=np.ones((2,3))
zeros_array=np.zeros((2,3))
print("\nOnes array:\n",ones_array)
print("\nZeros array:\n",zeros_array)
```

```python
random_6_by_3=np.random.rand(6,3);
transposed=random_6_by_3.T
print("\nRandom 6*3 array:\n",random_6_by_3)
print("\nTransposed array (3*6):\n",transposed)
```

```python
n,p,m=3,4,2
A=np.random.rand(n,p)
B=np.random.rand(p,m)
dot_product=np.dot(A,B)
print("\nMatrix A (n*p):\n",A)
print("\nMatrix B (p*m):\n",B)
print("\nDot Product (n*m):\n",dot_product)
```
